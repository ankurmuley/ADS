```python
########################################################################################################
#                          Assignment 1: Author, Vaishali Lambe NUID: 001286444                        #
########################################################################################################

#import pandas as pd to read data from ClassData.csv
import pandas as pd 

#read the dataset
ClassData=pd.read_csv("C:/Users/Admin/Documents/DataScience/R-tutorial/ClassData.csv")

#rename few colunms in dataframe ClassData
ClassData = ClassData.rename(columns={'Years of work experience': 'YWEx', 'Latest salary (per year)': 'LSPY',
                                      'Expected Salary after graduation': 'ESAG','Coops/Internships (YN)':'Coop',
                                      'Number of contacts on Linkedin' :'NCLN' })
#display the dataset in dataframe format
ClassData
```

```python
#Min, Max, Average, Median of GPA
#import numpy as np to caculate min max median and mean
import numpy as np
meanGPA=np.mean(ClassData['GPA'])
print("Average of GPA is:",meanGPA)
maxGPA=np.max(ClassData['GPA'])
print("Maximum value of GPA is:",maxGPA)
minGPA=np.min(ClassData['GPA'])
print("Minimum value of GPA is:",minGPA)
medGPA=np.median(ClassData['GPA'])
print("Median value of GPA is:",medGPA)
```

```python
#Min, Max, Average, Median of years of work experience
meanYWEx=np.mean(ClassData['YWEx'])
print("Average of years of work exp is:",meanYWEx)
maxYWEx=np.max(ClassData['YWEx'])
print("maximum vaue of years of work exp is:", maxYWEx)
minYWEx=np.min(ClassData['YWEx'])
print("minimum value of years of work exp is:", minYWEx)
medYWEx=np.median(ClassData['YWEx'])
print("madian of years of work exp is:", medYWEx)
```

```python
#mode of latest salary per year
#import stats from scipy 
from scipy import stats

#calculate mode of latest salary per year
modeLSPY=stats.mode(ClassData['LSPY'])
print("mode of latest salary per year is:", modeLSPY)
```

```python
#% of students having Co/op and not having Co/op
#total students
total_count=len(ClassData['Coop'])

#total students on coop
On_Coop=ClassData['Coop'].value_counts()['Y']
On_Coop1=ClassData['Coop'].value_counts()['y']
Total_on_Coop= On_Coop+On_Coop1

#total students not on coop
Non_Coop=ClassData['Coop'].value_counts()['N']

#percent of students having coop
percent_on_Coop= (Total_on_Coop)/(total_count) *100
print("% of students having co-op is:", percent_on_Coop)

#percent of students not having coop
percent_non_Coop= (Non_Coop)/(total_count) *100
print("% of students not having co-op is:", percent_non_Coop)
```

```python
# No of students with more than 500 LinkedIn contacts

ln_Contacts=ClassData[(ClassData['NCLN'] > 500 )]
print("No. of students with more than 500 contacts are:", len(ln_Contacts))
```

```python
# Find the Inter Quartile Range for the Expected
iqr = np.subtract(*np.percentile(ClassData['ESAG'], [75, 25]))
print("Inter Quartile Range for the Expected for expected salary is:", iqr)
```
